```python
import data as d
```

```python
data_list = d.get_data(5000, 16, 25, 10)
```

```python
import numpy as np
```

```python
fs = np.fft.rfft(data_list[11][4500:4550, 0, :].reshape(-1), norm='ortho')
```

```python
fs.real += (np.random.random(fs.shape) * 2 - 1) * abs(fs).max() * 0.1
fs.imag += (np.random.random(fs.shape) * 2 - 1) * abs(fs).max() * 0.1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fs.real, 'r-', fs.imag, 'b-')
```

```python
inverted = np.fft.irfft(fs, norm='ortho')
```

```python
plt.rcParams["figure.figsize"] = (40,5)
plt.plot(inverted, 'r-', data_list[11][4500:4550, 0, :].reshape(-1), 'g--')
```
